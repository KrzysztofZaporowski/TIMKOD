# Zadanie 1
### Treść 
Zadanie polega na wygenerowaniu tak zwanego przybliżenia zerowego rzędu dla języka angielskiego. Czyli ciągu losowych znaków (26 liter alfabetu + spacja), w którym symbole mają jednakową szansę na wystąpienie (1/27) niezależnie od poprzednio wygenerowanych symboli. Wskazówka: kolejne zadania również dotyczą generowania innych przybliżeń, zaprojektowanie kodu w oparciu o generator, który przyjmuje źródło informacji jako argument, może zaoszczędzić pisania sporej ilości podobnego kodu. Jaka jest średnia długość słowa w tym przybliżeniu? Wskazówka: podzielenie liczby znaków przez liczbę spacji w tekście nie jest poprawną metodą policzenia średniej długości słowa!

```python
def zero_order_generator(alphabet):
    while True:
        yield random.choice(alphabet)
```

```python
alphabet = string.ascii_lowercase + " "

gen = zero_order_generator(alphabet)
sample_text = "".join(next(gen) for _ in range(200))

print(f"Wygenerowany tekst:\n{sample_text}")
```

```python
words = sample_text.split()
words_count = len(words)
total_length = sum(len(word) for word in words)

if words_count > 0:
    print(f"Liczba słów: {words_count}")
    print(f"Łączna długość słów: {total_length}")
    print(f"Srednia długość słowa: {total_length / words_count:.2f}")
```

Dla tekstu z pliku `norm_wiki_sample.txt`

```python
with open("norm_wiki_sample.txt", "r") as f:
    wiki_text = f.read().lower()

gen_wiki = zero_order_generator(wiki_text)
sample_wiki_text = "".join(next(gen_wiki) for _ in range(200))

wiki_words = sample_wiki_text.split()
wiki_words_count = len(wiki_words)
total_length_wiki = sum(len(word) for word in wiki_words)

if wiki_words_count > 0:
    print(f"Liczba słów w norm_wiki_sample.txt: {wiki_words_count}")
    print(f"Łączna długość słów w norm_wiki_sample.txt: {total_length_wiki}")
    print(f"Srednia długość słowa w norm_wiki_sample.txt: {total_length_wiki / wiki_words_count:.2f}")
```

Jaka jest średnia długość słowa w tym przybliżeniu? Wskazówka: podzielenie liczby znaków przez liczbę spacji w tekście nie jest poprawną metodą policzenia średniej długości słowa!

Średnia długość słowa zależy od tego co wygeneruje generator, ale w przybliżeniu to średnia długość słowa to 26 ponieważ szansa na spacje wynosi $\frac{1}{27}$, a na litery $\frac{26}{27}$ co daje nam $\frac{\frac{26}{27}}{\frac{1}{27}} = 26$

### Treść 
Czy różne znaki występują jednakowo często w prawdziwym tekście? Oblicz częstość wystąpienia poszczególnych znaków w angielskim tekście. Wskazówka: nie musisz zliczać wszystkich znaków w korpusie, wystraczy policzyć na odpowiednio dużej próbce. Jakie znaki są najbardziej prawdopodobne, a które najmniej? Zwróć uwagę na kody przypisane poszczególnym literom w kodzie Morse’a, czy widzisz jakąś zależność, z czego wynika?

```python
letters = {
    'a': 0,
    'b': 0,
    'c': 0,
    'd': 0,
    'e': 0,
    'f': 0,
    'g': 0,
    'h': 0,
    'i': 0,
    'j': 0,
    'k': 0,
    'l': 0,
    'm': 0,
    'n': 0,
    'o': 0,
    'p': 0,
    'q': 0,
    'r': 0,
    's': 0,
    't': 0,
    'u': 0,
    'v': 0,
    'w': 0,
    'x': 0,
    'y': 0,
    'z': 0,
    ' ': 0
}
```

```python
def count_letters_frequency(file_name, letters_dict):
    with open(file_name, 'r') as file:
        text = file.read().lower()
    
    probably_valid_characters = {}

    for char in text:
        if char in letters_dict:
            letters_dict[char] += 1

    valid_characters_count = sum(letters_dict.values())
    print(f"Wszytskie litery: {valid_characters_count}")

    for char in letters_dict:
        probably_valid_characters[char] = letters_dict[char] / valid_characters_count
        print(f"{char} : {letters_dict[char]} : {(letters_dict[char] / valid_characters_count) * 100:.2f}%")

    sum_of_frequencies = 0
    for char in letters_dict:
        sum_of_frequencies += letters_dict[char] / valid_characters_count
    print(f"Sum of frequencies: {sum_of_frequencies:.4f}")

    return probably_valid_characters
```

```python
hamlet_probabilities = count_letters_frequency('norm_hamlet.txt', letters)
top5 = Counter(hamlet_probabilities).most_common(5)
keys = [item[0] for item in top5]
print(f"Najczęściej występujące litery w norm_hamlet.txt: {keys}")
bottom5 = Counter(hamlet_probabilities).most_common()[:-6:-1]
bottom5_keys = [item[0] for item in bottom5]
print(f"Najrzadziej występujące litery w norm_hamlet.txt: {bottom5_keys}")
```

```python
romeo_probabilities = count_letters_frequency('norm_romeo.txt', letters)
top5 = Counter(romeo_probabilities).most_common(5)
keys = [item[0] for item in top5]
print(f"Najczęściej występujące litery w norm_romeo.txt: {keys}")
bottom5 = Counter(romeo_probabilities).most_common()[:-6:-1]
bottom5_keys = [item[0] for item in bottom5]
print(f"Najrzadziej występujące litery w norm_romeo.txt: {bottom5_keys}")
```

```python
wiki_probabilities = count_letters_frequency('norm_wiki_sample.txt', letters)
top5 = Counter(wiki_probabilities).most_common(5)
keys = [item[0] for item in top5]
print(f"Najczęściej występujące litery w norm_wiki_sample.txt: {keys}")
bottom5 = Counter(wiki_probabilities).most_common()[:-6:-1]
bottom5_keys = [item[0] for item in bottom5]
print(f"Najrzadziej występujące litery w norm_wiki_sample.txt: {bottom5_keys}")
```

Związek z Kodem Morsa jest taki że litery które występują najczęściej takie jak: e, a, t, i mają najkrótsze kody, a te co występują najrzadziej mają najdłuższe kody

```python
def first_order_generator(probabilities_dict):
    characters = list(probabilities_dict.keys())
    probabilities = list(probabilities_dict.values())
    
    while True:
        yield random.choices(characters, weights = probabilities)[0]
```

# Zadanie 4
### Treść 
Czy różne znaki występują jednakowo często po sobie? Oblicz prawdopodobieństwo wystąpienia poszczególnych znaków po każdym z 2. najczęściej występujących znaków w korpusie.

```python
def calculate_pairs_frequencies(file_name, target_letters):
    with open(file_name, 'r', encoding='utf-8') as file:
        text = file.read().lower()

    stats = {char: Counter() for char in target_letters}

    for i in range(len(text) - 1):
        current_char = text[i]
        if current_char in stats:
            next_char = text[i+1]
            stats[current_char][next_char] += 1

    for char, counts in stats.items():
        total_occurrences = sum(counts.values())
        print(f"Pairs for '{char}':")
        print(f"Total occurrences followed by something: {total_occurrences}")

        for next_char, count in counts.most_common():
            freq_pct = (count / total_occurrences) * 100
            display_next = f"'{next_char}'" if next_char != " " else "' '"
            print(f"  '{char}{next_char}' : {count} : {freq_pct:.2f}%")
        print()
```

```python
# Hamlet pairs frequencies
probabilities = dict(sorted(hamlet_probabilities.items(), key=lambda item: item[1], reverse=True))
key_values = list(probabilities.keys())[0:4]

calculate_pairs_frequencies('norm_hamlet.txt', key_values)
```

```python
# Romeo sample text
probabilities = dict(sorted(romeo_probabilities.items(), key=lambda item: item[1], reverse=True))
key_values = list(probabilities.keys())[0:2]

calculate_pairs_frequencies('norm_romeo.txt', key_values)
```

```python
# Wiki sample text
probabilities = dict(sorted(wiki_probabilities.items(), key=lambda item: item[1], reverse=True))
key_values = list(probabilities.keys())[0:2]

calculate_pairs_frequencies('norm_wiki_sample.txt', key_values)
```

Na koniec wygeneruj przybliżenie źródła Markova piątego rzedu. Zacznij od ciągu znaków zawierającego już słowo “probability”. (4pt) Jaka jest średnia długość wyrazu w tych przybliżeniach?

```python
def count_mean_word_length(markov_output):
    words = markov_output.split()
    words_count = len(words)
    total_length = sum(len(word) for word in words)
    return total_length / words_count if words_count > 0 else 0
```

```python
def build_markov_model(text, order):
    model = defaultdict(Counter)
    for i in range(len(text) - order):
        context = text[i : i + order]
        next_char = text[i + order]
        model[context][next_char] += 1
    return model
```

```python
def generate_markov_text(model, order, length, start_context=None):
    if start_context is None or len(start_context) < order:
        context = random.choice(list(model.keys()))
        result = context
    else:
        result = start_context
        context = start_context[-order:]

    for _ in range(length):
        if context not in model:
            break
        
        choices = list(model[context].keys())
        weights = list(model[context].values())
        
        next_char = random.choices(choices, weights=weights)[0]
        result += next_char
        context = result[-order:] 
        
    return result
```
